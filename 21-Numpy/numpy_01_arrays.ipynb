# Guia de Introdução ao NumPy

NumPy, que significa Numerical Python, é uma biblioteca fundamental para a computação científica em Python. Ela fornece suporte para arrays e matrizes, além de funções matemáticas para operações com esses objetos. É, também, a base da biblioteca Pandas.



## Arrays

Um array é uma estrutura de dados que armazena valores do mesmo tipo. Em Python, isso é uma grande vantagem porque economiza espaço e permite operações mais eficientes. Vamos criar um array.



```python
import numpy as np

#criação de um array
array_bao = np.array(
    [1, 2, 3, 4, 5]
)

display(array_bao)
```

```python
array_bil = np.array(
    ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
)
array_bil
```

## Conhecendo um array: indexação e slice


```python
array_bao[-2]
```

```python
# slicing x:y -> do indice x até y-1
array_bil[1:4]
```

```python
array_bao[0:-1]
```

```python
array_bil[int(len(array_bil)/2):len(array_bil)]
```

```python
for i, e in enumerate(array_bao):
    print(f'{i} -> {e}')
```

```python
for i, e in enumerate(array_bil):
    print(f'{i} -> {e}')
```

```python
#indice inverso
array_bil[-3]
```

```python
#passo
# array [inicio:fim:passo]
array_bil[0:-1:2]
```

```python
for i, e in enumerate(array_bil[0:-1:2]):
    print(f'{i*2} -> {e}')
```

```python
array_bil[::]
```

```python
array_bil[5::2]
```

```python
array_bil[::-1]
```
