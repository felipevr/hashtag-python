### Puxar Base de dados

```python
import pandas as pd

caminho = '22-Pandas/Pandas 23-Kaggle/'


def formatar(valor):
    return 'US$ {:_.2f}'.format(valor).replace('.', ',').replace('_', '.')


tabela = pd.read_csv(caminho+"exportacoes_franca.csv")
display(tabela)
```

### Como foi a evolução das exportações para a frança ao longo dos anos?

Crie uma tabela com os anos de 2016, 2017, 2018, 2019 e 2020 e o valor US$ de exportações

```python
anos = tabela["Year"].unique()

tabela_anos_exp = tabela[["Year", 'US$ FOB']].groupby("Year").sum()

tabela_anos_exp2 = tabela_anos_exp['US$ FOB'].apply(formatar)

display(tabela_anos_exp2)

tabela_anos_exp.plot()
```

### Quais os produtos mais exportados ao longo de todo o período?

Queremos saber os produtos com maior valor US$ de exportação

```python
tabelas_produtos_exportados = tabela[["SH2 Description", "US$ FOB"]].groupby("SH2 Description").sum().sort_values(by="US$ FOB", ascending=False)

tabelas_produtos_exportados["US$ FOB"] = tabelas_produtos_exportados["US$ FOB"].apply(formatar)

tabelas_produtos_exportados
```

### Quais os produtos mais exportados (em US$) que as 2 maiores cidades (em exportação em 2020) exportaram?

```python
tabelas_produtos_2020 = tabela[tabela["Year"] == 2020]

tabelas_CIDADES_2020 = tabelas_produtos_2020[["City", "US$ FOB"]].groupby("City").sum().sort_values(by="US$ FOB", ascending=False)

tabelas_CIDADES_2020["US$ FOB"] = tabelas_CIDADES_2020["US$ FOB"].apply(formatar)

tabelas_CIDADES_2020
```

### Quais são os produtos mais exportados (em US$) que as 2 maiores cidades (em exportação em 2020) exportaram?

```python
cidade1 = "Duque de Caxias - RJ"
cidade2 = "Luís Eduardo Magalhães - BA"

def pega_produtos_mais_exportados_cidade(cidadeX):
    filtro = tabelas_produtos_2020["City"] == cidadeX
    tabela_cidadeX = tabelas_produtos_2020.loc[filtro, :]

    tabela_prod_cidadeX = tabela_cidadeX[["SH2 Description", "US$ FOB"]].groupby("SH2 Description").sum()

    tabela_prod_cidadeX = tabela_prod_cidadeX.sort_values(by="US$ FOB", ascending=False)

    tabela_prod_cidadeX["US$ FOB"] = tabela_prod_cidadeX["US$ FOB"].apply(formatar)

    return tabela_prod_cidadeX

print(f"Produto mais exportados da cidade: ", cidade1)
tabela_prod_duque_caxias = pega_produtos_mais_exportados_cidade(cidade1)
display(tabela_prod_duque_caxias)
```

```python
print(f"Produto mais exportados da cidade: ", cidade2)
tabela_prod_cidade2 = pega_produtos_mais_exportados_cidade(cidade2)
display(tabela_prod_cidade2)
```

```python
top_5_cidades = tabelas_CIDADES_2020.index[:5]
for cidade in top_5_cidades:
    print(f"Produto mais exportados da cidade: ", cidade)
    tabela_prod_cidade = pega_produtos_mais_exportados_cidade(cidade)
    display(tabela_prod_cidade[:1])
```
