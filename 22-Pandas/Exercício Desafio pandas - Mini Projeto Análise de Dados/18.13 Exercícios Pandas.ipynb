### 1. Valor Total da Folha Salarial -> Qual foi o gasto total com salários de funcionários pela empresa? <br>
    Sugestão: calcule o salário total de cada funcionário, salário + benefícios + impostos, depois some todos os salários

```python
salarios_base = df_funcionarios['Salario Base'].sum()
impostos = df_funcionarios['Impostos'].sum()
beneficios = df_funcionarios['Beneficios'].sum()
```

```python
df_funcionarios['Nova'] = df_funcionarios['Beneficios'] + df_funcionarios['VT'] + df_funcionarios['VR']
#df_funcionarios['Beneficios'] = df_funcionarios['Beneficios'].str.replace(',','.')
# df_funcionarios['Beneficios'].fillna(0)
# df_funcionarios['Beneficios'].astype(float)

extras = df_funcionarios['Nova'].sum()

total_folha = salarios_base + impostos + extras

total_folha_formatado = 'R$ {:_.2f}'.format(total_folha).replace('.', ',').replace('_', '.')
print("Valor Total Folha Salarial:", total_folha_formatado)
```

### 2. Qual foi o faturamento da empresa?<br>
    Sugestão: calcule o faturamento total de cada serviço e depois some o faturamento de todos

```python
import pandas as pd

df_servicos = pd.read_excel('22-Pandas/Exercício Desafio pandas - Mini Projeto Análise de Dados/BaseServiçosPrestados.xlsx')

df_servicos.head()
```

```python

df_clientes = pd.read_csv('22-Pandas/Exercício Desafio pandas - Mini Projeto Análise de Dados/CadastroClientes.csv', sep=';', decimal=',')

df_clientes.head()
```

```python
junta_df = df_clientes.merge(df_servicos) #.groupby(['ID CLiente', 'CLiente']).sum(
junta_df
```

```python
junta_df["Valor Cobrado"] = junta_df["Valor Contrato Mensal"] * junta_df["Tempo Total de Contrato (Meses)"]
junta_df
```

```python
faturamento = junta_df["Valor Cobrado"].sum()

faturamento_formatado = 'R$ {:_.2f}'.format(faturamento).replace('.', ',').replace('_', '.')
print("Valor Faturamento da Empresa:", faturamento_formatado)
```

### 3. Qual o % de funcionários que já fechou algum contrato?
<br>
    Sugestão: na base de serviços temos o funcionário que fechou cada serviço. Mas nem todos os funcionários que a empresa tem já fecharam algum serviço.<br>

    . Queremos calcular Qtde_Funcionarios_Fecharam_Serviço / Qtde_Funcionários_Totais<br>
    . Para calcular a qtde de funcionários que fecharam algum serviço, use a base de serviços e conte quantos funcionários tem ali. Mas lembre-se, cada funcionário só pode ser contado uma única vez.<br><br>
    Dica: se você aplicar o método .unique() em uma variável que é apenas 1 coluna de um dataframe, ele vai excluir todos os valores duplicados daquela coluna.<br>
    Ex: unicos_colunaA = dataframe['colunaA'].unique() te dá como resposta uma lista com todos os itens da colunaA aparecendo uma única vez. Todos os valores repetidos da colunaA são excluidos da variável unicos_colunaA 

```python
len(df_servicos["ID Funcionário"].unique())
```

```python
catraca = df_servicos.groupby(by="ID Funcionário").sum("Tempo Total de Contrato (Meses)")
catraca
```

```python
percentual_funcionarios = len(catraca) / len(df_funcionarios)
#percentual_funcionarios
percentual_formatado = '{:.2%}'.format(percentual_funcionarios).replace('.', ',').replace('_', '.')
print(f"{percentual_formatado} funcionários fecharam contrato com algum cliente.")
```



### 6. Qual o ticket médio mensal (faturamento médio mensal) dos contratos?<br>
    Dica: .mean() calcula a média -> exemplo: media_colunaA = dataframe['colunaA'].mean()

```python
nova_junta = junta_df.drop(['ID Cliente', 'Cliente', 'ID Funcionário'], axis=1)
nova_junta.head()
```

```python
nova_junta["Media"] = nova_junta["Valor Cobrado"] / nova_junta["Tempo Total de Contrato (Meses)"]
nova_junta.head()
```

```python
media = nova_junta["Media"].mean()

media_formatado = 'R$ {:_.2f}'.format(media).replace('.', ',').replace('_', '.')
print("Ticket Médio Mensal dos contratos:", media_formatado)
```
