# Comparando, Tratando e Mesclando DataFrames

## Objetivo

Vamos modificar os IDs para os nomes dos produtos, dos clientes e das lojas, para nossas análises ficarem mais intuitivas futuramente. Para isso, vamos criar um data frame com todos os detalhes.

- Usaremos o método merge para isso e, depois se quisermos, podemos pegar apenas as colunas que queremos do dataframe final.

### Criando nossos dataframes

```python
import pandas as pd

#às vezes precisaremos mudar o encoding. Possiveis valores para testar:
#encoding='latin1', encoding='ISO-8859-1', encoding='utf-8' ou então encoding='cp1252'
vendas_df = pd.read_csv(r'22-Pandas\Contoso - Vendas - 2017.csv', sep=';')


#usaremos o display para ver todos os dataframes
display(vendas_df.head())
```

```python
produtos_df = pd.read_csv(r'22-Pandas\Contoso - Cadastro Produtos.csv', sep=';', encoding='utf-8')
produtos_df.head()
```

```python
lojas_df = pd.read_csv(r'22-Pandas\Contoso - Lojas.csv', sep=';')
lojas_df.head()
```

```python

clientes_df = pd.read_csv(r'22-Pandas\Contoso - Clientes.csv', sep=';')

promocoes_df = pd.read_csv(r'22-Pandas\Contoso - Promocoes.csv', sep=';')

display(clientes_df.head())
display(promocoes_df[:10])
```

### Vamos tirar as colunas inúteis do clientes_df ou pegar apenas as colunas que quisermos

```python
# axis=1 -> indica que quer remover colunas e não linhas
promocoes_df = promocoes_df.drop(['Unnamed: 5','Unnamed: 6'], axis=1)
promocoes_df.head()
```

```python
# axis=1 -> indica que quer remover colunas e não linhas
clientes_df = clientes_df.drop(['Unnamed: 7','Unnamed: 8','Unnamed: 9','Unnamed: 10'], axis=1)
clientes_df.head()
```

```python
clientes_df2 = clientes_df[['ID Cliente', 'E-mail']]
display(clientes_df2.head())

produtos_df2 = produtos_df[['ID Produto', 'Nome do Produto']]
display(produtos_df2.head())

lojas_df2 = lojas_df[['ID Loja', 'Nome da Loja']]
display(lojas_df2.head())

promocoes_df2 = promocoes_df[['ID Promocao', 'Nome Promocao']]
display(promocoes_df2.head())
```

### Agora vamos juntar os dataframes para ter 1 único dataframe com tudo "bonito"

- Obs: O merge precisa das colunas com o mesmo nome para funcionar. Se não tiver, você precisa alterar o nome da coluna com o .rename

```python
#juntando os dataframes
vendas_df = vendas_df.merge(produtos_df2, on='ID Produto')
vendas_df = vendas_df.merge(clientes_df2, on='ID Cliente')
vendas_df = vendas_df.merge(lojas_df2, on='ID Loja')
vendas_df = vendas_df.merge(promocoes_df2, on='ID Promocao')


#exibindo o dataframe final
vendas_df
```

```python
#vamos renomear o e-mail para ficar claro que é do cliente
vendas_df = vendas_df.rename(columns={'E-mail':'E-mail do Cliente'})
vendas_df.head()
```
