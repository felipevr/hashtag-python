```python
import pandas as pd

caminho = '00-playground/'

# Ler os arquivos CSV em dataframes separados
df1 = pd.read_csv(caminho+'parametros_integracao_siscad_202404121841.csv')
df2 = pd.read_csv(caminho+'parametros_monitoramento_tutores_checkout_202404121841.csv')
```

```python
# Mesclar os dataframes com base na coluna 'moodle_course_id'
df_merged_in = pd.merge(df1, df2, on='moodle_course_id', how='right')

df_merged_in.to_csv(caminho+'blah.csv')

# Verificar o dataframe mesclado
display(df_merged_in)
```

```python
# Mesclar os dataframes com base na coluna 'moodle_course_id'
df_merged_out = pd.merge(df1, df2, on='moodle_course_id', how='outer')

#df_merged_out['moodle_course_id'] = df_merged_out['moodle_course_id'].astype(int)

# Verificar o dataframe mesclado
display(df_merged_out)
```

```python
# Suponha que df1 e df2 sejam seus dataframes
# Concatenar os dataframes
concatenated_df = pd.concat([df_merged_in, df_merged_out])

display(concatenated_df)

# Remover duplicatas mantendo apenas a primeira ocorrência
# Isso resultará em um dataframe contendo apenas as linhas distintas
unique_rows = concatenated_df.drop_duplicates(keep=False)

display(unique_rows)
```
