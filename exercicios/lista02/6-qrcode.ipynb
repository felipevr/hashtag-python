## QR Code com imagem

````python
import qrcode
import qrcode.constants
from qrcode.image.styledpil import StyledPilImage

caminho = 'exercicios/lista02/'

def gerar_qrcode_imagem(nome, url):

    qr = qrcode.QRCode(error_correction=qrcode.constants.ERROR_CORRECT_H)
    qr.add_data(url)

    image = qr.make_image(
        image_factory=StyledPilImage,
        embeded_image_path=caminho+"logo.png"
    )
    #display(type(image), type(qr))

    image.save(f'{caminho}qrcode_{nome}.png')

    return image

image = gerar_qrcode_imagem('logo', 'https://www.youtube.com/@HashtagProgramacao')

image
````

## Criando diversos QR Code ao mesmo tempo

````python
redes_sociais = {
    'Facebook':'https://www.facebook.com/hashtagprogramacao',
    'Instagram':'https://www.instagram.com/hashtagprogramacao',
    'Youtube':'https://www.youtube.com/@HashtagProgramacao',
    'Tiktok': 'https://www.tiktok.com/@hashtagprogramacao'
}

for rede, url in redes_sociais.items():
    #print(rede, url)
    gerar_qrcode_imagem(rede, url)
````
